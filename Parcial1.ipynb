```python

import numpy as np
import matplotlib.pyplot as plt

x,step=np.linspace(0.1,1.1,num=101,retstep=True)
f=lambda x: np.sqrt(np.tan(x))
df_real=lambda x:(1/np.cos(x))**2/(2*np.sqrt(np.tan(x)))

def derivada_progresiva(x,f,h=step):
    return (     (1/(2*h))*(-3*f(x)+4*f(x+h)-f(x+2*h))                )

def derivada_central(x,f,h=step):
    return (f(x+h)-f(x-h))/(2*h)

df_p=derivada_progresiva(x,f)
df_c=derivada_central(x,f)
```

```python

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.scatter(x,df_p,color='m')
ax.set_title('Derivada progresiva')
ax.set_xlabel('X')
ax.set_ylabel("f'(X)")
plt.show()
```

```python


fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.scatter(x,df_c,color='y')
ax.set_title('Derivada central')
ax.set_xlabel('X')
ax.set_ylabel("f'(X)")
plt.show()
```

```python



fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(x,df_real(x),'k',label="D'REAL")
ax.scatter(x,df_c,label="D'Central",color='y')
ax.scatter(x,df_p,label="D'Progresiva",color='m')
ax.legend()
ax.set_title('Derivada central')
ax.set_xlabel('X')
ax.set_ylabel("f'(X)")
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.scatter(x,abs(df_real(x)-df_c),label="D'Central",color='y')
ax1.scatter(x,abs(df_real(x)-df_p),label="D'Progresiva",color='m')
ax1.legend()
ax1.set_title('Error')
ax1.set_xlabel('X')
ax1.set_ylabel('Error')
plt.show()
```
